```
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (40,15)
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)
#plot frequency of items python
def plotFreq(word_list,numResults):

  counts = Counter(word_list)
  labels, values = zip(*counts.items())

  # sort your values in descending order
  indSort = np.argsort(values)[::-1]

  # rearrange your data
  labels = np.array(labels)[indSort][:numResults]
  values = np.array(values)[indSort][:numResults]

  indexes = np.arange(len(labels))

  bar_width = 0.35

  plt.bar(indexes, values)

  # add labels
  plt.xticks(indexes + bar_width, labels)
  plt.show()
  print(len(labels))
  print(len(values))
  """
  for f, b in zip(labels, values):

    print(f, b)
  """
word_list = ['A', 'A', 'B', 'B', 'A', 'C', 'C', 'C', 'C']
numResults=3
plotFreq(word_list,numResults)
```
